```python
import pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Revenue Growth Factor: 18% of prior period Revenue
EBITDA Margin: 45.0% of current Revenue
Taxes: 21% of current EDBITDA
Capital Expenditures: 3% of current Revenue
Working Capital Increase: 6% of Revenue
EBITDA Exit Multiple: 32.70
Discount Factor: 0.94 (6%). 

```python
years = pandas.date_range(start ='2022',end ='2032', freq='Y').year
years
```

```python
df =pd.DataFrame(columns = ["Revenue", "EBITDA", "Taxes","Capital Expenditure", "Working Capital","Cash Flow", "Discount Factor",
                            "Present Value Free Cash Flow"]
                 , index= years)
df.loc[2022,"Revenue"] = 222 * (1 + 0.18)
df
```

```python

for i in range (1,len(df)):
    df.iloc[i]["Revenue"] = df.iloc[i -1]["Revenue"] * (1 + 0.18)
```

```python
df
```

```python
df["EBITDA"] =df ["Revenue"] * 0.45
```

```python
df["Taxes"] = -df ["EBITDA"] * 0.21
```

```python
df["Capital Expenditure"] = -df ["Revenue"] * 0.03
```

```python
df["Working Capital"] = -df ["Revenue"] * 0.06 
```

```python
df["Cash Flow"] = df["EBITDA"] + df["Taxes"] + df["Capital Expenditure"] + df["Working Capital"]
```

```python
df
```

```python
df.iloc[0]["Discount Factor"]= 0.94
for i in range(1,len(df)):
  
    df.iloc[i]["Discount Factor"] = (1 + 0.06) ** -(i +1)
```

```python
df
```

```python
df["Present Value Free Cash Flow"] = df["Cash Flow"] * df["Discount Factor"]
```

```python
df
```

```python
Revenue_2032 = df.loc[2031,"Revenue"]* (1.18)
Revenue_2032
```

```python
EBITDA_2032 = Revenue_2032 * 0.45
EBITDA_2032
```

```python
terminal_value = EBITDA_2032 * 32.70
terminal_value
```

```python
present_teminal_value = terminal_value * df.loc[2031, "Discount Factor"]
present_teminal_value
```

```python
present_value_of_projected = df["Cash Flow"].sum()
present_value_of_projected
```
